### Практическая часть.

#### Есть выборка из 12 квартир. Известны их площади (x) и стоимости (y). Построить уравнение парной линейной регрессии, оценить ее точность и адекватность. С помощью уравнения регрессии посчитать значения, предсказанные моделью значения цен на квартиры.

```python
import numpy as np
```

```python
n = 12
```

```python
x = np.array([80, 90, 85, 115, 85, 85, 90, 80, 105, 110, 65, 100], dtype=np.float64)
y = np.array([150, 160, 155, 175, 140, 150, 140, 155, 165, 190, 140, 165], dtype=np.float64)
```

#### Для нахождения коэффициентов регрессии $a$ и $b$ воспользуемся приведенными выше формулами:

```python
b = (np.mean(x * y) - np.mean(x) * np.mean(y)) / (np.mean(x**2) - np.mean(x) ** 2)
b
```

```python
a = np.mean(y) - b * np.mean(x)
a
```

#### Уравнение регрессии:  

```python
# С увеличением площади квартиры на 1 квадратный метр цена возрастет на 0.87 тысячи долларов.
```

#### Найдем коэффициент корреляции $r$ с помощью коэффициента $b$ и средних квадратического отклонения, посчитанного для массивов $x$ и $y$:

```python
r = b * np.std(x) / np.std(y)
r
```

#### Найдем коэффициент детерминации $R^2$:

```python
R2 = r**2
R2
```

```python
# Это означает, что 67.5% вариации цены на квартиру (y) объясняется вариацией фактора x — площади квартиры.
```

#### посчитаем значения, предсказанные моделью значения цен на квартиры:

```python
y_pred = a + b * x
y_pred
```

#### Качество модели найдем с помощью средней ошибки аппроксимации $\overline {A}$:

```python
A_mean = 100 * np.mean(np.abs((y - y_pred) / y))
A_mean
```

#### Так как $\overline {A}$ равна 4%, что не превышает 8-10 %, модель хорошо описывает эмпирические данные. 

#### Для оценки значимости уравнения регрессии воспользуемся F-критерием Фишера. Найдем фактическое значение $F$-критерия ($F_{факт}$):

```python
F_fact = (r**2 * (n - 2)) / (1 - r**2)
F_fact
```

#### При 5 % уровне значимости и степенях свободы $k_1 = 1$ и $k_2 = 12 - 2 = 10$ табличное значение критерия: $F_{кр} = 4.96$. Так как $F_{факт} = 20.79 > F_{кр} = 4.96$, уравнение регрессии статистически значимо.

#### Для оценки статистической значимости параметров регрессии воспользуемся $t$-статистикой Стьюдента и также рассчитаем доверительные интервалы каждого из показателей. При $df = n - 2 = 12 - 2 = 10$ и $\alpha = 0.05$ получим: 

#### Определим стандартную ошибку $S_{ост}$ (переменная **s_residual**) и случайные ошибки $m_a, \; m_b$:

```python
s_residual = np.sqrt(np.sum((y - y_pred)**2) / (n - 2))
m_a = s_residual * np.sqrt(np.sum(x ** 2)) / (n * np.std(x))
m_b = s_residual / (np.std(x) * np.sqrt(n))
print('s_residual = {}\nm_a = {}\nm_b = {}'.format(s_residual, m_a, m_b))
```

#### Вычислим наблюдаемые значения критерия $t_a$ и $t_b$:

```python
t_a = a / m_a
t_a
```

```python
t_b = b / m_b
t_b
```

#### Фактические значения t-статистики больше табличного значения: 

#### поэтому параметры  a  и  b  не случайно отличаются от нуля, то есть они статистически значимы

#### Рассчитаем доверительные интервалы для параметров регрессии $a$ и $b$. Для этого определим предельную ошибку для каждого показателя ($\Delta_a$ и $\Delta_b$), используя значение $t_{кр}$, равное 2.228 (переменная **t_cr**):

```python
t_cr = 2.228
delta_a = t_cr * m_a
delta_a
```

```python
delta_b = t_cr * m_b
delta_b
```

#### Найдем границы доверительных интервалов $\gamma_{a_{min}}, \gamma_{a_{max}}, \gamma_{b_{min}}, \gamma_{b_{max}}$:

```python
gamma_a_max = a + delta_a
gamma_a_max
```

```python
gamma_a_min = a - delta_a
gamma_a_min
```

```python
gamma_b_max = b + delta_b
gamma_b_max
```

```python
gamma_b_min = b - delta_b
gamma_b_min
```

#### Приходим к выводу о том, что с вероятностью $p = 1 - \alpha = 0.95$ параметры $a$ и $b$, находясь в указанных границах, являются статистически значимыми и отличны от нуля. Поместим исходные и предсказанные данные в датафрейм **df**:

```python
import pandas as pd
df = pd.DataFrame({'x':x, 'y':y, 'y_pred':y_pred}, columns=['x', 'y', 'y_pred'])
df
```

```python
df = df.sort_values('x')
df
```

#### Построим на одном графике исходные данные и теоретическую прямую, построенную по уравнению регрессии:

```python
import matplotlib.pyplot as plt
plt.scatter(df['x'], df['y'])
plt.plot(df['x'], df['y_pred'])
plt.xlabel('Площадь квартиры')
plt.ylabel('Цена квартиры')
plt.show()
```
